# **Функции перевода из 10ой СС: bin(), oct() и hex().**
---
**Функция bin()** используется для преобразования целого числа в строку двоичного формата. Форматированная строка имеет префикс «0b».

Функция **oct()** используется для преобразования целого числа в восьмеричную строку с префиксом «0o».

Функция **hex()** используется для преобразования целого числа в шестнадцатеричную строку в нижнем регистре с префиксом «0x».

**Тип возвращаемого значения - строка.**

```python
a=12

#-------------функции перевода из 10ой СС-----------------
bin_a=bin(a)                # функция bin() переводит из 10ой СС в 2ую СС
print(bin_a, type(bin_a))   # 0b1100 <class 'str'>
oct_a=oct(a)                # функция oct() переводит из 10ой СС в 8ую СС
print(oct_a, type(oct_a))   # 0o14 <class 'str'>
hex_a=hex(a)                # функция hex() переводит из 10ой СС в 16ую СС
print(hex_a, type(hex_a))   # 0xc <class 'str'>
```

# **Функции перевода из любой СС в 10ую СС: int().**

---


Функция **int()** принимает два параметра. Первый — строка, которую нужно преобразовать в число. Второй — система счисления, в которой представлено число. Второй параметр не обязателен.

```python
#-------------функция перевода в 10ую СС - int()-----------
print(int(bin_a,2))         # 12
print(int(oct_a,8))         # 12
print(int(hex_a,16))        # 12
print(int('204',5))         # 54 = 4+0*5+2*5**2
print(int('1100',2))        # 12
```

# **Работа со строками**


---

**Индексирование строк  и срезы в Python**

**Индексирование** строк в Python представляет собой процесс получения отдельных символов из строки по их позиции в строке. Индексы строк в Python начинаются с 0, где первый символ имеет индекс 0, второй — индекс 1, и так далее. Можно также использовать отрицательные индексы, где -1 соответствует последнему символу в строке, -2 — предпоследнему, и так далее.

С помощью **срезов** можно получать части строки на основе их позиций в строке, что позволяет извлекать подстроки с определенными индексами или диапазонами индексов.

**Синтаксис** срезов: **s[start:end:step]** где
    
    start — начальный индекс среза (включительно),
    end — конечный индекс среза (исключительно), и
    step — шаг среза (опционально).

```python
#-------------Работа со строками. Срезы.-------------------
s='0123456789'
print(len(s))            # len() возвращает длину строками
print(s[0],s[1])         # строка - массив из символов. Нумерация начинается с нуля
print(s[-1],s[-2])       # выведет 9 8
print(s[2:])             # срез начиная с индекса 2 до конца строки (по умолчанию)
print(s[:5])             # срез начиная с индекса 0 (по умолчанию) до индекса 5 не включительно
print(s[1:7:3])          # срез от индекса 1 до индекса 7 (правая граница не включительная) с шагом 3
```

# **Строки. Функции и методы строк**

---

**len(S)** - функция возвращает длину строки.

**S.count(str, [start],[end])** - возвращает количество непересекающихся вхождений подстроки в диапазоне [начало, конец] (0 и длина строки по умолчанию).

**S.replace(шаблон, замена[, maxcount])** - заменяет шаблон на замену, maxcount ограничивает количество замен.

**S.find(str, [start],[end])** - поиск подстроки в строке. Возвращает номер первого вхождения или -1.

**S.zfill(width)** - делает длину строки не меньшей width, по необходимости заполняя первые символы нулями.

**S.strip([chars])** - удаление пробельных символов в начале и в конце строки.

**S.lstrip([chars])** - удаление пробельных символов в начале строки.

**S.rstrip([chars])** - удаление пробельных символов в конце строки.

```python
#-------------Работа со строками. Функции.----------------
print(bin_a.count('1'))         # 2. Функция count подсчитывает количество символов 1
print(bin_a.replace('0','2'))   # 1122. Функция replase заменяет 0 на 2
print(bin_a.replace('0','2',1)) # 1120. Функция replase заменяет 0 на 2 один раз
print(bin_a.zfill(8))           # 00001100. Функция zfill делает заданную строку не меньше указанной длины,
                                # по необходимости дополняя первые символы нулями
print(bin_a.find('0'))          # 2. Функция find возвращает индекс первого вхождения подстроки
                                # если не найдена, то возвращает -1
a='0001010000'

print(a.strip('0'))             # 101. Удаляет начальные и конечные символы
print(a.lstrip('0'))            # 1010000. Удаляет начальные символы
print(a.rstrip('0'))            # 000101. Удаляет конечные символы
```

Списки в Python - упорядоченные изменяемые коллекции объектов произвольных типов. Список может содержать любое количество любых объектов (в том числе и вложенные списки), или не содержать ничего.
Используете список, когда необходимо подсчитать сколько одинаковых элементов получилось после выполнения алгоритма.

list = [] - создает пустой список c именем list
list.append(x) - добавляет элемент x в конец списка.
list.count(x) - возвращает количество элементов со значением x.

Множества — это неупорядоченная коллекция уникальных элементов, сгруппированных под одним именем. Используйте множество, когда необходимо найти количество уникальных элементов.

_set = set() - создает пустое множество с именем _set.
_set.add(x) - добавляет элемент x во множество.
len(_set) - подсчитывает количество элементов, входящих во множество.



```python
#-------------Структуры данных, которые могут пригодиться--
list = []                       # список
list.append(5)                  # добавляет элемент в конец списка
print(list,len(list))           # [5] 1
                                # используйте список, когда необходимо узнать количесво чисел, которые удовлетворяют условию
_set = set()                    # множество - это контейнер, содержащий не повторяющиеся элементы
_set.add(3)
_set.add(3)
_set.add(3)
print(_set,len(_set))           # {3} 1
                                # используйте множество, когда необходимо найти количесво уникальных элементов
```
